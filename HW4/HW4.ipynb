```python
#Importing libraries
#bheng2 bhoyer2 NETID
import pandas as pd
import numpy as np
from sympy import *
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
HW4 = 'IE420.s2022.HW4.data.xlsx'
```

```python
# Question 0 begin
fund_names = pd.read_excel(HW4, sheet_name='Fund names', header= 1)
fund_navs = pd.read_excel(HW4, sheet_name='Fund NAVs', header= 1)
returns = pd.read_excel(HW4, sheet_name='returns', header= 1)

#fund_names.head()
#fund_navs.head()
#returns.head()
returns
```

```python
#Add NaN for -99 and remove unused data
returns = returns.where(returns != -99)
returns.drop(returns.index[0:24], inplace = True)
returns
```

```python
#Calculations and additions for fund excess returns
excess_returns = pd.DataFrame()
excess_returns = excess_returns.assign(FBCVX_r = returns['FBCVX'] - returns['rf'])
excess_returns = excess_returns.assign(FCPVX_r = returns['FCPVX'] - returns['rf'])
excess_returns = excess_returns.assign(FDGRX_r = returns['FDGRX'] - returns['rf'])
excess_returns = excess_returns.assign(FGRTX_r = returns['FGRTX'] - returns['rf'])
excess_returns = excess_returns.assign(FLCSX_r = returns['FLCSX'] - returns['rf'])
excess_returns = excess_returns.assign(FSCRX_r = returns['FSCRX'] - returns['rf'])
excess_returns = excess_returns.assign(market_r = returns['rM - rf'])
print('q0) Dataframe with fund and market excess returns:')
print(excess_returns)
```

```python
# Question 1 beginning

FBCVXr = excess_returns['FBCVX_r'].values.reshape(-1)
FCPVXr = excess_returns['FCPVX_r'].values.reshape(-1)
FDGRXr = excess_returns['FDGRX_r'].values.reshape(-1)
FGRTXr = excess_returns['FGRTX_r'].values.reshape(-1)
FLCSXr = excess_returns['FLCSX_r'].values.reshape(-1)
FSCRXr = excess_returns['FSCRX_r'].values.reshape(-1)
rSMB = returns['rSMB'] 
rHML = returns['rHML']
rSMBr = rSMB.values.reshape(-1)
rHMLr = rHML.values.reshape(-1)
rM_rf = returns['rM - rf'].values.reshape(-1)


FBCVXr_filtered = FBCVXr[np.logical_not(np.isnan(FBCVXr))]
FCPVXr_filtered = FCPVXr[np.logical_not(np.isnan(FCPVXr))]
FSCRXr_filtered = FSCRXr[np.logical_not(np.isnan(FSCRXr))]
```

```python
# Question 1a

FBCVX_beta, intercept, r_value, p_value, strder = st.linregress(rM_rf[len(rM_rf)-len(FBCVXr_filtered):], FBCVXr_filtered)
FCPVX_beta, intercept, r_value, p_value, strder = st.linregress(rM_rf[len(rM_rf)-len(FCPVXr_filtered):],FCPVXr_filtered)
FDGRX_beta, intercept, r_value, p_value, strder = st.linregress(rM_rf,FDGRXr)
FGRTX_beta, intercept, r_value, p_value, strder = st.linregress(rM_rf,FGRTXr)
FLCSX_beta, intercept, r_value, p_value, strder = st.linregress(rM_rf,FLCSXr)
FSCRX_beta, intercept, r_value, p_value, strder = st.linregress(rM_rf[len(rM_rf)-len(FSCRXr_filtered):],FSCRXr_filtered)


print('q1a) Beta for funds:')
print('FBCVX_beta', 'FCPVX_beta', 'FDGRX_beta', 'FGRTX_beta', 'FLCSX_beta', 'FSCRX_beta')
Beta = np.array([FBCVX_beta, FCPVX_beta, FDGRX_beta, FGRTX_beta, FLCSX_beta, FSCRX_beta])

#Beta = pd.DataFrame(Beta_arr.reshape(1,6), columns = ['FBCVX_beta', 'FCPVX_beta', 'FDGRX_beta', 'FGRTX_beta', 'FLCSX_beta', 'FSCRX_beta'])

#Beta_arr = {'FBCVX_beta' : FBCVX_beta, 'FCPVX_beta' : FCPVX_beta, 'FDGRX_beta' : FDGRX_beta, 'FGRTX_beta' : FGRTX_beta,'FLCSX_beta' : FLCSX_beta, 'FSCRX_beta':FSCRX_beta }
#Beta = pd.DataFrame(data = Beta_arr)
#FBCVX_beta, FCPVX_beta, FDGRX_beta, FGRTX_beta, FLCSX_beta, FSCRX_beta]
#['FBCVX_beta', 'FCPVX_beta', 'FDGRX_beta', 'FGRTX_beta', 'FLCSX_beta', 'FSCRX_beta']
print(Beta)
print()

Alpha_final1 = (FBCVXr_filtered.mean() - (Beta[0]*rM_rf[len(rM_rf)-len(FBCVXr_filtered):].mean()))
Alpha_final2 = (FCPVXr_filtered.mean() - (Beta[1]*rM_rf[len(rM_rf)-len(FCPVXr_filtered):].mean()))
Alpha_final3 = (excess_returns["FDGRX_r"].mean() - (Beta[2]*excess_returns["market_r"].mean()))
Alpha_final4 = (excess_returns["FGRTX_r"].mean() - (Beta[3]*excess_returns["market_r"].mean()))
Alpha_final5 = (excess_returns["FLCSX_r"].mean() - (Beta[4]*excess_returns["market_r"].mean()))
Alpha_final6 = (excess_returns["FSCRX_r"].mean() - (Beta[5]*excess_returns["market_r"].mean()))

Alpha_final = {Alpha_final1, Alpha_final2, Alpha_final3, Alpha_final4, Alpha_final5, Alpha_final6}
alphasfinal = pd.DataFrame(Alpha_final, columns = ['Alpha Final Values'], index = ['FCPVX','FBCVX','FLCSX','FDGRX','FGRTX','FSCRX'])
alphasfinal

#index = ['FCPVX','FBCVX','FLCSX','FDGRX','FGRTX','FSCRX']

print(Alpha_final1, Alpha_final2, Alpha_final3, Alpha_final4, Alpha_final5, Alpha_final6)
print('q1a) Alpha for funds:')
print(alphasfinal)
print()
print('q1a) Average of the 6 alpha values:', alphasfinal.mean())
```

```python
# Question 1b

a,b = st.t.interval(alpha=0.95, df = len(alphasfinal) - 1, loc = np.mean(alphasfinal), scale = st.sem(alphasfinal))
print()
print('q1b) Confidence Interval bounds:')
print(a,b)

ConfidenceAboveb = alphasfinal.where(alphasfinal > b)
ConfidenceBelowa = alphasfinal.where(alphasfinal < a)
print('q1b) Threshold for significant 5% threshold upper bound:')
print(ConfidenceAboveb)
print('q1b) Threshold for significant 5% threshold lower bound:')
print(ConfidenceBelowa)
```

```python
#Slicing out data from rows we don't use (NaNs)

#len(FBCVXr_filtered)
#len(FCPVXr_filtered)
market_r_1 = rM_rf[len(rM_rf)-len(FBCVXr_filtered):]
market_r_2 = rM_rf[len(rM_rf)-len(FCPVXr_filtered):]
rSMB_1 = rSMB[len(rSMB)-len(FBCVXr_filtered):]
rSMB_2 = rSMB[len(rSMB)-len(FCPVXr_filtered):]
rHML_1 = rSMB[len(rHML)-len(FBCVXr_filtered):]
rHML_2 = rSMB[len(rHML)-len(FCPVXr_filtered):]
```

```python
#Question 2a

Alpha1 = FBCVXr_filtered.mean() - BetaM1 * market_r_1.mean() - BetaSMB1 * rSMB_1.mean() - BetaHML1 * rHML_1.mean()
Alpha2 = FCPVXr_filtered.mean() - BetaM2 * market_r_2.mean() - BetaSMB2 * rSMB_2.mean() - BetaHML2 * rHML_2.mean()
Alpha3 = excess_returns['FDGRX_r'].mean() - BetaM3 * excess_returns['market_r'].mean() - BetaSMB3 * rSMB.mean() - BetaHML3 * rHML.mean()
Alpha4 = excess_returns['FGRTX_r'].mean() - BetaM4 * excess_returns['market_r'].mean() - BetaSMB4 * rSMB.mean() - BetaHML4 * rHML.mean()
Alpha5 = excess_returns['FLCSX_r'].mean() - BetaM5 * excess_returns['market_r'].mean() - BetaSMB5 * rSMB.mean() - BetaHML5 * rHML.mean()
Alpha6 = excess_returns['FSCRX_r'].mean() - BetaM6 * excess_returns['market_r'].mean() - BetaSMB6 * rSMB.mean() - BetaHML6 * rHML.mean()

Alphas = {Alpha1, Alpha2, Alpha3, Alpha4, Alpha5, Alpha6}
alphas = pd.DataFrame(Alphas, index = ['FLCSX','FCPVX','FDGRX','FSCRX','FGRTX','FBCVX'], columns = ['Alpha Values 2'])
alphas
```
